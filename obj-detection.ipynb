```python
import yaml

data_yaml = {
    "train": "/kaggle/input/yolo-dataset/yolo_dataset/images/train",
    "val": "/kaggle/input/yolo-dataset/yolo_dataset/images/val",
    "test": "/kaggle/input/yolo-dataset/yolo_dataset/images/test",
    "nc": 1,  # Change to actual number of classes
    "names": ["Person"]  # Change to actual class names
}

with open("data.yaml", "w") as f:
    yaml.dump(data_yaml, f)

print("‚úÖ data.yaml created!")
```

```python
# Install YOLOv8

import ultralytics
from ultralytics import YOLO
import torch

# Check if GPU is available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Dataset Path
dataset_path = "/kaggle/input/yolo-dataset"

# Train YOLOv8 Model
model = YOLO("yolov8n.yaml")  # Use 'yolov8n.yaml' for a small model or 'yolov8s.yaml' for a larger one

model.train(
    data="data.yaml",
    epochs=50,
    batch=16,
    imgsz=640,
    device=device,
    patience=3  # Stops if no improvement for 5 epochs
)



# Save trained model
model.export(format="onnx")  # Export to ONNX format for deployment
model_path = "yolov8_trained.pt"
print(f"‚úÖ Model trained and saved as {model_path}")

```

```python
metrics = model.val(data="data.yaml")  # Validate on the dataset
print(metrics)  # Show evaluation results
```

```python
# Print mean Average Precision (mAP) at IoU 0.50 and 0.50-0.95
print(f"mAP@50: {metrics.box.map50:.3f}")  
print(f"mAP@50-95: {metrics.box.map:.3f}")

# Extract precision and recall (mean of all classes)
precision = sum(metrics.box.p) / len(metrics.box.p)  # Mean Precision
recall = sum(metrics.box.r) / len(metrics.box.r)    # Mean Recall

print(f"Precision: {precision:.3f}")
print(f"Recall: {recall:.3f}")
```

```python
import glob
import cv2
from ultralytics import YOLO

# Load the trained YOLOv8 model
model = YOLO("runs/detect/train/weights/best.pt")  

# Path to the folder containing test images
test_images_path = "/kaggle/input/yolo-dataset/yolo_dataset/images/test/*.jpg"  

# Get all image file paths
image_files = glob.glob(test_images_path)

# Run inference on each image
for image_path in image_files:
    results = model(image_path, conf=0.5)  

    # Process the first result
    result = results[0]

    # Load image for displaying results
    img = cv2.imread(image_path)

    # Draw bounding boxes and labels
    for box in result.boxes:
        x1, y1, x2, y2 = map(int, box.xyxy[0])  # Bounding box coordinates
        confidence = box.conf[0].item()  # Confidence score
        
        # Draw bounding box
        cv2.rectangle(img, (x1, y1), (x2, y2), (255, 0, 0), 2)

        # Set label as "person" instead of "object"
        label = f"person {confidence:.2f}"
        cv2.putText(img, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

    # Save the modified image
    output_path = image_path.replace("test", "output")  # Save in output folder
    cv2.imwrite(output_path, img)
    print(f"‚úÖ Inference complete for {image_path}. Saved as {output_path}")

print("‚úÖ All images processed successfully!")
```
